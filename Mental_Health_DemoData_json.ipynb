```python
# adding a column called diagnosis which will populate the name of the diagnosis found in each record
#create an empty list to hold name of diagnosis 

diagnosis_list = []

for i, row in clean_mh_demo_data_df.iterrows():
    diagnosis_list = []
    if row['ADHDFLG'] == 1:
        diagnosis_list.append('Attention deficit/hyperactivity disorder (ADHD)')
    if row['ALCSUBFLG'] == 1:
        diagnosis_list.append('Alcohol or substance use disorders')
    if row['ANXIETYFLG'] == 1:
        diagnosis_list.append('Anxiety disorders')
    if row['BIPOLARFLG'] == 1:
        diagnosis_list.append('Bipolar disorders')
    if row['CONDUCTFLG'] == 1:
        diagnosis_list.append('Conduct disorders')
    if row['DELIRDEMFLG'] == 1:
        diagnosis_list.append('Delirium/dementia')     
    if row['DEPRESSFLG'] == 1:
        diagnosis_list.append('Depressive disorders')
    if row['ODDFLG'] == 1:
        diagnosis_list.append('Oppositional defiant disorders')
    if row['PDDFLG'] == 1:
        diagnosis_list.append('Pervasive developmental disorder')
    if row['PERSONFLG'] == 1:
        diagnosis_list.append('Personality disorders')
    if row['SCHIZOFLG'] == 1:
        diagnosis_list.append('Schizophrenia or other psychotic disorders')
    if row['TRAUSTREFLG'] == 1:
        diagnosis_list.append('Trauma- and stressor-related disorders')
    if row['OTHERDISFLG'] == 1:
        diagnosis_list.append('Other disorders/conditions') 
    if not diagnosis_list:
        diagnosis_list.append("Information unavailable")
        
    diagnosis_list.append(diagnosis_list)

        
```

```python
clean_mh_demo_data_df.at[i, "DIAGNOSIS"] = diagnosis_list
```
