```python
def get_diagnosis(row):
    diagnoses = []
    if row['ADHDFLG'] == 1:
        diagnoses.append('Attention deficit/hyperactivity disorder (ADHD)')
    if row['ALCSUBFLG'] == 1:
        diagnoses.append('Alcohol or substance use disorders')
    if row['ANXIETYFLG'] == 1:
        diagnoses.append('Anxiety disorders')
    if row['BIPOLARFLG'] == 1:
        diagnoses.append('Bipolar disorders')
    if row['CONDUCTFLG'] == 1: 
        diagnoses.append('Conduct disorders')
    if row['DELIRDEMFLG'] == 1:
        diagnoses.append('Delirium/dementia')     
    if row['DEPRESSFLG'] == 1:
        diagnoses.append('Depressive disorders')
    if row['ODDFLG'] == 1:
        diagnoses.append('Oppositional defiant disorders')
    if row['PDDFLG'] == 1:
        diagnoses.append('Pervasive developmental disorder')
    if row['PERSONFLG'] == 1:
        diagnoses.append('Personality disorders')
    if row['SCHIZOFLG'] == 1:
        diagnoses.append('Schizophrenia or other psychotic disorders')
    if row['TRAUSTREFLG'] == 1:
        diagnoses.append('Trauma- and stressor-related disorders')
    if row['OTHERDISFLG'] == 1:
        diagnoses.append('Other disorders/conditions') 
    if not diagnoses:
        diagnoses.append("Information unavailable")
    return ", ".join(diagnoses)

diagnosis_data_df = diagnosis_data_df.assign(DIAGNOSIS=diagnosis_data_df.apply(get_diagnosis, axis=1))
```

```python
#Then the query() method is used to filter the rows based on specific conditions.
#The function get_diagnosis takes in a row and returns the diagnosis based on the column values. 
diagnosis_data_df = diagnosis_data_df.assign(DIAGNOSIS = diagnosis_data_df.query("ADHDFLG == 1 or ALCSUBFLG == 1 or ANXIETYFLG == 1 or BIPOLARFLG == 1 or CONDUCTFLG == 1 or DELIRDEMFLG == 1 or DEPRESSFLG == 1 or ODDFLG == 1 or PDDFLG == 1 or PERSONFLG == 1 or SCHIZOFLG == 1 or TRAUSTREFLG == 1 or OTHERDISFLG == 1").apply(get_diagnosis, axis=1))
diagnosis_data_df = diagnosis_data_df.drop(columns=["ADHDFLG", "ALCSUBFLG", "ANXIETYFLG", "BIPOLARFLG", "CONDUCTFLG", "DELIRDEMFLG", "DEPRESSFLG", "ODDFLG", "PDDFLG", "PERSONFLG", "SCHIZOFLG", "TRAUSTREFLG", "OTHERDISFLG"])
```
